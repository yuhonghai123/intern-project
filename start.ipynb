```python
import requests
from bs4 import BeautifulSoup
import re

def extract_chinese(text):
    pattern = re.compile('[\u4e00-\u9fa5，。；！？“”：；‘’~……]+')
    chinese_text = ' '.join(pattern.findall(text))
    return chinese_text

def get_article_content(url):
    try:
        # 发送GET请求获取网页内容
        response = requests.get(url)
        response.raise_for_status()  # 检查请求是否成功
        # 使用BeautifulSoup解析HTML
        # 解决乱码问题
        response.encoding=response.apparent_encoding
        soup = BeautifulSoup(response.text, 'html.parser')
        print(soup)
        # 根据网站的HTML结构定位文章主要内容
        # 请根据实际网站结构进行修改
        article_content = soup.find('title').get_text()
        article_content = extract_chinese(str(soup))

        return article_content

    except requests.exceptions.RequestException as e:
        print(f"Error: {e}")
        return None

# 使用示例
article_url = 'https://auto.eastday.com/404.html'
content = get_article_content(article_url)

if content:
    print("Article Content:")
    print(content)
else:
    print("Failed to retrieve article content.")
```
